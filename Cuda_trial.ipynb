```python
# 3. Write CUDA code in a cell using the %%cuda magic
%%cuda
#include <stdio.h>

__global__ void add(float *a, float *b, float *c) {
    int tid = threadIdx.x;
    c[tid] = a[tid] + b[tid];
}

int main() {
    printf("CUDA code executed successfully!\n Hello World");
    return 0;
}
```
